```

#Importo las librerias necesarias en el codigo:
import numpy as np
import matplotlib.pyplot as plt
import math



def eulerMet(ti,tn,yj,n):
    """Funcion que almacena las variables
    iniciales, una lista de ceros para definir
    las dimensiones ,y un espacio lineal."""
    h = 0.20 #Elijo 0.20 pero cualquier valor de paso es posible
    t = np.linspace(ti,tn,n+1)
    y = np.zeros(n+1)
    y[0]= yj
  
  #Utilizo un ciclo for para el despeje, luego grafico la solucion
  #que ya conozco de la EDO para su comparacion/verificacion:
    for j in range(1,n+1):
        y[j]= (y[j-1]+20*h*math.cos(t[j-1]+h)-h*math.sin(t[j-1]+h))/(1+20*h)
        y2[j]= (-math.exp(-20*t[j-1])+math.cos(t[j-1]))
    return ((t,y,y2))

#Por ultimo coloco las lineas capaz de graficar la solucion y la respuesta conocida.

print("Euler hacia atras")
(t,y1,y2) = eulerMet(0,2,0,200)
plt.plot(t,y1,label = "Euler Hacia atras")
plt.plot(t,y2,label="Soluci√≥n exacta")
plt.legend()

#Comparamos
```
